```python
#Ronald Davidson
#December 13, 2022
#Final Project

#Project description: This project uses a lot of the algorithms that we looked at in class this semester in order to display the appropriate information.
#Most of the choices will use execption handling functions to prevent the user from triggering a run-time error. The results for each choice are displayed
#Using print statements which can be found in the main function of the program.

#Basic function to open the data file
def openFile():
     goodFile = False
     while goodFile == False:
        fname = input("Enter the name of the data file: ")
        try:
            inFile = open(fname, 'r')
            goodFile = True
        except IOError:
            print("Invalid file name - try again")
     return inFile

#Function reads lines in data file to return the information stored in the file into lists
#Parameters: None
# Returns titleList, ratingList, yearList, scoreList
def getData():
    inFile = openFile()
    titleList = []
    ratingList = []
    yearList = []
    scoreList = []
    line = inFile.readline()
    for line in inFile:
        line = line.strip()
        title,rating,year,score = line.split(',')
        titleList.append(title)
        ratingList.append(rating)
        yearList.append(int(year))
        scoreList.append(score)
    inFile.close()
    return titleList, ratingList, yearList, scoreList
#Exeption handling for correct choice
#Paramters: None
#Returns choice
def correctChoice():
    validInt = False
    while validInt == False:
        try:
             #Asks user for choice
            choice = int(input("Choice ==> "))
            #If choice is not between 1-7 return invalid
            if choice >= 1 and choice <=7:
                 validInt = True
                 return choice
            print("Invalid entry - try again")
        except ValueError:
             print("Invalid entry - try again")
    return choice

#Ask the user what choice they want.
#Parameters: None
#Returns choice
def getChoice():
    print("")
    print("Please choose one of the following options:")
    print("1 -- Find all shows with a certain rating")
    print("2 -- Find the show with the highest score released in a specified range of years")
    print("3 -- Search for a show by title") 
    print("4 -- Find the average score for films with a specific rating")
    print("5 -- Find all shows with a score higher than the score for a given show")
    print("6 -- Sort all lists by year and write results to a new file")
    print("7 -- Quit")
    choice =correctChoice()
    print("")
    return choice

# choice 1

#Function asks for the correct rating that the user wants to search for
#Parameters: None
#returns the rating
def correctRating(ratingList):
    validInt = False
    while validInt == False:
        try:
            Rating =(str(input("Enter rating: ")))
            if  Rating in ratingList:
                validInt = True
            else:
                 print ("Invalid entry - try again")
        except ValueError:
            print ("Invalid entry - try again")
    return Rating


#choice 2
#Functionmakes sure user entered year is valid
#Parameters: YearList and a string
#Returns year
def getYear(yearList, string):
    x = 0
    goodValue = False
    while goodValue == False:
        try:
            year = int(input(string))
            if int(year) >= 2005 and int(year) <= 2017:
     
                 goodValue = True
            else:
                 print("Invalid year - try again")
        except ValueError:
            print ("Invalid entry - try again")
    return year

#Function checcks to see if the second user inputted year is greater than the first year
#Paramters: yearList
#Returns second rating
def checkYears(yearList):
     print("Enter year range to search (oldest year first)")
     bad = True
     while bad == True:
          year1 = getYear(yearList, "Year1: ")
          year2 = getYear(yearList, "Year2: ")
          if year2 < year1:
               print("Second year should be after first year - try again")
          else:
               bad = False
     return year1, year2

#Function will look through all list of titles and append the ones that fit the time slot into a new list
# Parameters: titleList, ratingList, yearList, scoreList
#Returns a list called findRightShows contaiing only titles of shows in the time slot
def findRightShows(titleList, ratingList, yearList, scoreList, year1, year2):
    findGoodScore = []
    findGoodTitle = []
    findGoodYear = []
    findGoodRating = []
    for i in range(len(yearList)):
         if int(yearList[i]) >= year1 and int(yearList[i]) <= year2:
              findGoodScore.append(scoreList[i])
              findGoodTitle.append(titleList[i])
              findGoodYear.append(yearList[i])
              findGoodRating.append(ratingList[i])
    return findGoodScore, findGoodTitle, findGoodYear, findGoodRating
    

#choice 3
#Function will check list to see if show is valid
#Parameters: titleList, ratingList, yearList, scoreList
#Returns nothing
def printResults(titleList, ratingList, yearList, scoreList):
    show = (str(input("Enter show title: ")))
    for i in range(len(titleList)):
         # str makes search non-case sensitive
         if titleList[i] == str.title(show):
              print("")
              print("The TV shows that meet your criteria are: ")
              print("")
              print("TITLE".ljust(40), "RATING".ljust(8), "YEAR".ljust(5), "SCORE".ljust(4))
              print(titleList[i].ljust(40), ratingList[i].ljust(8),str(yearList[i]).ljust(5),str(scoreList[i]).ljust(4))
              return
    print("")
    print("No shows meet your criteria")
    return
    
#choice 4
#Function will search through titleList and will add the scores of all titles with the correct rating to claculate average.
#Parameters: titleList, ratingList, yearList, scoreList
#Returns average and Rating
def averageScore(ratingList, scoreList, Rating):
    averageList = []
    average = 0
    for i in range(len(ratingList)):
        if ratingList[i] == Rating:
            averageList.append(scoreList[i])
    for i in range(len(averageList)):
        average = average + int(averageList[i])
    average = average/len(averageList)

    return average, Rating


#choice 5

#Function uses exception handling to make sure a show is valid
#Parameters: titleList
#Returns show
def getValidShow(titleList):
    validInt = False
    while validInt == False:
        show = (input("Enter title:"))
        try:
             
            for i in range(len(titleList)):
                 if show == titleList[i]:
                      validInt = True
                      return show
            print ("Invalid entry - try again")
        except ValueError:
             print ("Invalid entry - try again")
    return show

     
#Function asks user for the name of the show they want to search and then finds the score of that show by searching through titleList until the correct score is found
#Paramters: titleList, scoreList
#Returns viewerScore
def getShowRating(titleList, scoreList, show):
    print("")
    i = 0
    while titleList[i] !=show:
        i = i + 1
    viewerScore = scoreList[i]
    return viewerScore

    

#choice 6
#Creates a list of indexs equal to the size of the yearList
#Parameters: yearList
#Returns index
def createIndexList(yearList):
     index = []
     for i in range(len(yearList)):
          index.append(i)
     return index

#Function sorts through the yearList. If the years are switched, the correspsonding positions in index are swapped as well
#Parameters: yearList, index
#Returns originalYearList, yearList, index
def dualSort(yearList, index):
    originalYearList = []
    originalYearList = yearList.copy()
    for i in range(1, len(yearList)):
         # Multiple saves for each list being sorted
        save1 = yearList[i]
        save2 = index[i]
        j = i
        while j > 0 and int(yearList[j - 1]) > int(save1):
            yearList[j] = yearList[j - 1]
            index[j] = index[j - 1]
            j = j - 1
        yearList[j] = save1    
        index[j] = save2
    return originalYearList,yearList,index
     
 
#Function will write the data with the new sorted lists
#Parameters: titleList, ratingList, yearList, scoreList, index, originalYearList
#Returns nothing. Writes the data file
def writeData(titleList, ratingList, yearList, scoreList, index, originalYearList):
     outname =("year-sorted-shows.csv")
     print("")
     print("Data sorted by years written to file", outname)
     outFile = open(outname, 'w')
     outFile.write("Title" + ',' + "Rating" + ',' + "Year"+ ',' + "Score" + '\n')
     for i in range (len(titleList)):
          outFile.write(str(titleList[index[i]]) + ',' + str(ratingList[index[i]]) + ',' + str(originalYearList[index[i]]) + ',' + str(scoreList[index[i]]) + '\n')
     outFile.close()
     return

#Function executes entire program
#Main gets data from file and stores it into lists. Then asks user for their input choice. The following function srun depending on what number choice the user inputs. If user inputs anything esle besides numbers 1-6, the program will display a good-bye message and shut down.
def main():
    titleList, ratingList, yearList, scoreList = getData()
    choice = getChoice()
    while choice != 7:
        #if user inputs choice 1 then this code will run
        if choice == 1:
            Rating = correctRating(ratingList)
            print("")
            print("The TV shows that meet your criteria are: ")
            print("")
            print("TITLE".ljust(40), "RATING".ljust(8), "YEAR".ljust(5), "SCORE".ljust(4))
            for i in range(len(ratingList)):
                 if ratingList[i] == Rating:
                      print(titleList[i].ljust(40), ratingList[i].ljust(8),str(yearList[i]).ljust(5),str(scoreList[i]).ljust(4))
            choice = getChoice()
        #if user inputs choice 2 then this code will run
        elif choice == 2:
            year1, year2 = checkYears(yearList)
            findGoodScore, findGoodTitle, findGoodYear, findGoodRating = findRightShows(titleList, ratingList, yearList, scoreList, year1, year2)
            highScore = 0
            j = 0
            for i in range(len(findGoodScore)):
                 if int(findGoodScore[i]) > int(highScore):
                      highScore = int(findGoodScore[i])
                      j = (i)
            print("")
            print("The TV shows that meet your criteria are: ")
            print("")
            print("TITLE".ljust(40), "RATING".ljust(8), "YEAR".ljust(5), "SCORE".ljust(4))
            print(findGoodTitle[j].ljust(40), findGoodRating[j].ljust(8),str(findGoodYear[j]).ljust(5),str(findGoodScore[j]).ljust(4))
            choice = getChoice()
        #if user inputs choice 3 then this code will run
        elif choice == 3:
            printResults(titleList, ratingList, yearList, scoreList)
            choice = getChoice()
         #if user inputs choice 4 then this code will run
        elif choice == 4:
            rating = correctRating(ratingList)
            average, Rating = averageScore(ratingList, scoreList, rating)
            print("The average score for shows with a", Rating, "rating is", "{:,.2f}".format(average))
            choice = getChoice()
         #if user inputs choice 5 then this code will run
        elif choice == 5:
            show = getValidShow(titleList)
            viewerScore = getShowRating(titleList, scoreList, show)
            if int(viewerScore) == 99:
                 print("No shows meet your criteria")
            else:
         
                 print("The TV shows that meet your criteria are:")
                 print("")
                 print("TITLE".ljust(40), "RATING".ljust(8), "YEAR".ljust(5), "SCORE".ljust(4))
            for i in range(len(ratingList)):
                 if int(viewerScore) < int(scoreList[i]):
                      print(titleList[i].ljust(40), ratingList[i].ljust(8),str(yearList[i]).ljust(5),str(scoreList[i]).ljust(4))
            print("")
            choice = getChoice()
         #if user inputs choice 6 then this code will run
        elif choice == 6:
            index = createIndexList(yearList)
            originalYearList, yearList, index = dualSort(yearList, index)
            writeData(titleList, ratingList, yearList, scoreList, index, originalYearList)
            choice = getChoice()
        else:
            choice = getChoice()
    print("Good-bye")
```
